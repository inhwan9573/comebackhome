```python
import math
import numpy as np
def dist(a, b):
    return math.sqrt(math.pow(a[0]-b[0],2) + math.pow(a[1]-b[1],2))
```

```python
centroids = [ (2, 5), (8, 3), (4,5) ]
x = (0, 8)
```

```python
centroid_distances = []
for centroid in centroids:
    centroid_distances.append(dist(x,centroid))
print(centroid_distances)
# argmin 최소값의 인덱스
print(np.argmin(centroid_distances))
```
